Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Menampilkan persentase data untuk kolom kategorikal dalam bentuk pie chart
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for column in categorical_columns:
    plt.figure(figsize=(6, 6))
    df[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
    plt.title(f'Percentage of {column.capitalize()}')
    plt.ylabel('')  # Menghilangkan label y untuk kejelasan
    plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_pie_chart(column_name, color_map='Pastel1'):
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    plt.figure(figsize=(7, 7))

    # Get the colormap
    cmap = plt.cm.get_cmap(color_map)

    # Get colors from the colormap based on the number of categories
    num_colors = len(labels)
    colors = cmap(np.linspace(0, 1, num_colors))

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14, color='navy')
    plt.show()

# Menggunakan berbagai color map untuk variasi
plot_pie_chart('gender', 'Pastel1')
plot_pie_chart('race/ethnicity', 'Set3')
plot_pie_chart('lunch', 'Spectral')
plot_pie_chart('parental level of education', 'Pastel2')
plot_pie_chart('test preparation course', 'Accent')
````

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Mengatur ukuran dan gaya untuk semua plot
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")

# Membuat plot untuk pengaruh `gender` terhadap nilai
plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='math score', data=df, palette="Wistia")
plt.title("Math Score by Gender")

plt.subplot(2, 3, 2)
sns.boxplot(x='gender', y='reading score', data=df, palette="Wistia")
plt.title("Reading Score by Gender")

plt.subplot(2, 3, 3)
sns.boxplot(x='gender', y='writing score', data=df, palette="Wistia")
plt.title("Writing Score by Gender")

# Membuat plot untuk pengaruh `race/ethnicity` terhadap nilai
plt.subplot(2, 3, 4)
sns.boxplot(x='race/ethnicity', y='math score', data=df, palette="OrRd")
plt.title("Math Score by Race/Ethnicity")

plt.subplot(2, 3, 5)
sns.boxplot(x='race/ethnicity', y='reading score', data=df, palette="OrRd")
plt.title("Reading Score by Race/Ethnicity")

plt.subplot(2, 3, 6)
sns.boxplot(x='race/ethnicity', y='writing score', data=df, palette="OrRd")
plt.title("Writing Score by Race/Ethnicity")

plt.tight_layout()
plt.show()

# Plot tambahan untuk `parental level of education`, `lunch`, dan `test preparation course`
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# Parental level of education
sns.boxplot(ax=axes[0, 0], x='parental level of education', y='math score', data=df, palette="YlOrBr")
axes[0, 0].set_title("Math Score by Parental Level of Education")
sns.boxplot(ax=axes[0, 1], x='parental level of education', y='reading score', data=df, palette="YlOrBr")
axes[0, 1].set_title("Reading Score by Parental Level of Education")
sns.boxplot(ax=axes[0, 2], x='parental level of education', y='writing score', data=df, palette="YlOrBr")
axes[0, 2].set_title("Writing Score by Parental Level of Education")

# Lunch
sns.boxplot(ax=axes[1, 0], x='lunch', y='math score', data=df, palette="YlGnBu")
axes[1, 0].set_title("Math Score by Lunch")
sns.boxplot(ax=axes[1, 1], x='lunch', y='reading score', data=df, palette="YlGnBu")
axes[1, 1].set_title("Reading Score by Lunch")
sns.boxplot(ax=axes[1, 2], x='lunch', y='writing score', data=df, palette="YlGnBu")
axes[1, 2].set_title("Writing Score by Lunch")

# Test preparation course
sns.boxplot(ax=axes[2, 0], x='test preparation course', y='math score', data=df, palette="BuPu")
axes[2, 0].set_title("Math Score by Test Preparation Course")
sns.boxplot(ax=axes[2, 1], x='test preparation course', y='reading score', data=df, palette="BuPu")
axes[2, 1].set_title("Reading Score by Test Preparation Course")
sns.boxplot(ax=axes[2, 2], x='test preparation course', y='writing score', data=df, palette="BuPu")
axes[2, 2].set_title("Writing Score by Test Preparation Course")

plt.tight_layout()
plt.show()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
import pandas as pd

# Memuat dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Menambahkan kolom baru untuk status kelulusan (Pass atau Fail) pada setiap mata pelajaran
df['math_status'] = df['math score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['reading_status'] = df['reading score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['writing_status'] = df['writing score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')

# Menghitung jumlah siswa yang Pass dan Fail untuk masing-masing pelajaran
math_counts = df['math_status'].value_counts()
reading_counts = df['reading_status'].value_counts()
writing_counts = df['writing_status'].value_counts()

# Menampilkan hasil
print("Math Pass/Fail Counts:")
print(math_counts)
print("\nReading Pass/Fail Counts:")
print(reading_counts)
print("\nWriting Pass/Fail Counts:")
print(writing_counts)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Mengatur ukuran dan gaya visualisasi
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")

# Membuat plot untuk pengaruh `gender` terhadap nilai
plt.subplot(3, 3, 1)
sns.boxplot(x='gender', y='math score', data=df, palette="pastel")
plt.title("Math Score by Gender")

plt.subplot(3, 3, 2)
sns.boxplot(x='gender', y='reading score', data=df, palette="pastel")
plt.title("Reading Score by Gender")

plt.subplot(3, 3, 3)
sns.boxplot(x='gender', y='writing score', data=df, palette="pastel")
plt.title("Writing Score by Gender")

# Membuat plot untuk pengaruh `race/ethnicity` terhadap nilai
plt.subplot(3, 3, 4)
sns.boxplot(x='race/ethnicity', y='math score', data=df, palette="Set3")
plt.title("Math Score by Race/Ethnicity")

plt.subplot(3, 3, 5)
sns.boxplot(x='race/ethnicity', y='reading score', data=df, palette="Set3")
plt.title("Reading Score by Race/Ethnicity")

plt.subplot(3, 3, 6)
sns.boxplot(x='race/ethnicity', y='writing score', data=df, palette="Set3")
plt.title("Writing Score by Race/Ethnicity")

# Membuat plot untuk pengaruh `parental level of education` terhadap nilai
plt.subplot(3, 3, 7)
sns.boxplot(x='parental level of education', y='math score', data=df, palette="pastel")
plt.xticks(rotation=45)
plt.title("Math Score by Parental Level of Education")

plt.subplot(3, 3, 8)
sns.boxplot(x='parental level of education', y='reading score', data=df, palette="pastel")
plt.xticks(rotation=45)
plt.title("Reading Score by Parental Level of Education")

plt.subplot(3, 3, 9)
sns.boxplot(x='parental level of education', y='writing score', data=df, palette="pastel")
plt.xticks(rotation=45)
plt.title("Writing Score by Parental Level of Education")

plt.tight_layout()
plt.show()

# Visualisasi tambahan untuk `lunch` dan `test preparation course`
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

sns.boxplot(ax=axes[0, 0], x='lunch', y='math score', data=df, palette="Blues")
axes[0, 0].set_title("Math Score by Lunch")

sns.boxplot(ax=axes[0, 1], x='lunch', y='reading score', data=df, palette="Blues")
axes[0, 1].set_title("Reading Score by Lunch")

sns.boxplot(ax=axes[0, 2], x='lunch', y='writing score', data=df, palette="Blues")
axes[0, 2].set_title("Writing Score by Lunch")

sns.boxplot(ax=axes[1, 0], x='test preparation course', y='math score', data=df, palette="Purples")
axes[1, 0].set_title("Math Score by Test Preparation Course")

sns.boxplot(ax=axes[1, 1], x='test preparation course', y='reading score', data=df, palette="Purples")
axes[1, 1].set_title("Reading Score by Test Preparation Course")

sns.boxplot(ax=axes[1, 2], x='test preparation course', y='writing score', data=df, palette="Purples")
axes[1, 2].set_title("Writing Score by Test Preparation Course")

plt.tight_layout()
plt.show()
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Memuat dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Menambahkan kolom baru untuk rata-rata nilai
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Menampilkan rata-rata nilai dari seluruh siswa
print("Rata-rata nilai seluruh mata pelajaran dari setiap siswa:")
print(df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'average_score']].head())

# Visualisasi korelasi rata-rata nilai dengan kategori menggunakan boxplot
plt.figure(figsize=(15, 10))

# Korelasi dengan gender
plt.subplot(3, 2, 1)
sns.boxplot(x='gender', y='average_score', data=df, palette="pastel")
plt.title("Average Score by Gender")

# Korelasi dengan race/ethnicity
plt.subplot(3, 2, 2)
sns.boxplot(x='race/ethnicity', y='average_score', data=df, palette="Set3")
plt.title("Average Score by Race/Ethnicity")

# Korelasi dengan parental level of education
plt.subplot(3, 2, 3)
sns.boxplot(x='parental level of education', y='average_score', data=df, palette="pastel")
plt.xticks(rotation=45)
plt.title("Average Score by Parental Level of Education")

# Korelasi dengan lunch
plt.subplot(3, 2, 4)
sns.boxplot(x='lunch', y='average_score', data=df, palette="Blues")
plt.title("Average Score by Lunch")

# Korelasi dengan test preparation course
plt.subplot(3, 2, 5)
sns.boxplot(x='test preparation course', y='average_score', data=df, palette="Purples")
plt.title("Average Score by Test Preparation Course")

plt.tight_layout()
plt.show()
````
